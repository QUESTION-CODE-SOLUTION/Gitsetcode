```python
AB=np.array([1,2,6,4,7,8])
print(AB)
DC=AB.reshape([3,2])# we are reshaping it , and remember valid row and column pair
print(DC)
```

```python
a=np.arange(24)
print(a)
print(a.ndim)#we are  giving dimensions to it
```

```python
b=a.reshape(2,4,3)
print(b)
print(b.ndim)
```
